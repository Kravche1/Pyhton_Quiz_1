```python
import pandas as pd
```

```python
# Load the data
data = pd.read_csv("survey_results_public.csv")
```

```python
# Extract desired rows and columns
desired_data = data[['ResponseId', 'MainBranch', 'Age', 'Employment', 'ConvertedCompYearly']]
```

```python
# Assign indexing to a specific column
desired_data.set_index('ResponseId', inplace=True)
```

```python
# Create a filter based on 2 parameters
filter_data = desired_data[(desired_data['Age'] == '25-34 years old') & (desired_data['Employment'] == 'Employed, full-time')]
print(filter_data)
```

```python
# Sort the table using the 2 parameters
sorted_data = desired_data.sort_values(by=['Age', 'ConvertedCompYearly'], ascending=[True, False])
print(sorted_data)
```

```python
# Use statistical functions
mean_salary = desired_data['ConvertedCompYearly'].mean()
std_dev_salary = desired_data['ConvertedCompYearly'].std()
median_salary = desired_data['ConvertedCompYearly'].median()
min_salary = desired_data['ConvertedCompYearly'].min()
max_salary = desired_data['ConvertedCompYearly'].max()

print("Mean Salary:", mean_salary)
print("Standard Deviation of Salary:", std_dev_salary)
print("Median Salary:", median_salary)
print("Minimum Salary:", min_salary)
print("Maximum Salary:", max_salary)
```

```python
import matplotlib.pyplot as plt

# Example of Bar Chart
age_counts = desired_data['Age'].value_counts()
age_counts.plot(kind='bar')
plt.title('Age Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

# Example of Line Chart
salary_trend = desired_data.groupby('Age')['ConvertedCompYearly'].mean()
salary_trend.plot(kind='line')
plt.title('Average Salary Trend by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Salary')
plt.show()
```
